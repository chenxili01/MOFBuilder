```python
from mofbuilder import MetalOrganicFrameworkBuilder as MofBuilder
import veloxchem as vlx

smi_str = {"bdc": "O=C([O-])C(C=C1)=CC=C1C([O-])=O"}
```

## Original nocap MOF-5 (2,2,2)

```python
mof = MofBuilder(mof_family="MOF-5")
mof.ostream.mute()
smi_str = {"bdc": "O=C([O-])C(C=C1)=CC=C1C([O-])=O"}
mol = vlx.Molecule.read_smiles(smi_str["bdc"])
mof.linker_molecule = mol
mof.node_metal = "Zn"
mof.termination = False
mof.supercell = (2,2,2)
nocap = mof.build()
nocap.show(residue_indices=True)
nocap.show()
```

### Cap MOF-5 (2,2,2)

```python
mof = MofBuilder(mof_family="MOF-5")
mof.ostream.mute()
mol = vlx.Molecule.read_smiles(smi_str["bdc"])
mof.linker_molecule = mol
mof.node_metal = "Zn"
mof.supercell = (2,2,2)
cap = mof.build()
cap.show(residue_indices=True)
cap.show()
```

## Cap MOF-5 (2,2,2) slab


```python
mof = MofBuilder(mof_family="MOF-5")
mof.ostream.mute()
mol = vlx.Molecule.read_smiles(smi_str["bdc"])
mof.linker_molecule = mol
mof.node_metal = "Zn"
mof.supercell = (2,2,2)
mof.supercell_custom_fbox = ([0.2,0.5],[0,2],[0,2])
slabcap = mof.build()
slabcap.show(residue_indices=True)
slabcap.show()
```

## Replace

```python
newmol = vlx.Molecule.read_smiles("O=C([O-])C(C=C1N)=C(N)C=C1C([O-])=O")
newmol.show()
slabcap.ostream.mute()
rp_slab = slabcap.replace(replace_indices=[824,816,334,804,796,344],new_linker_molecule=newmol)
rp_slab.show()
```

## Remove

```python
nocap.update_node_termination = True
nocap.ostream.mute()
rm_a=nocap.remove(remove_indices=[333,223,231,117,7,79,189,1,73,183,295,111,405,217,289,327,399])
rm_a.show(residue_indices=False)
```
